```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
emes = [200, 350, 500, 750, 1000, 1250, 1500, 1750, 2000]
fallas = ['NoFault', 'AB', 'BC', 'CA', 'ABC', 'AG', 'BG', 'CG']
distancias = [10, 20, 50, 75, 95]

falla_true = ['BC' for x in range(10)]
dist_true = [95 for x in range(10)]

delays = []
accs_faults = []
accs_dists = []

for m in emes:
    df = pd.read_excel(f'BC_95/delays_m{m}.xlsx')

    falla_pred = df['Falla'].values
    dist_pred = df['Distancia'].values
    delay = np.average(df['Delay'])
    delays.append(delay)

    accuracy_fallas = accuracy_score(falla_true, falla_pred)
    accs_faults.append(accuracy_fallas)

    accuracy_dists = accuracy_score(dist_true, dist_pred)
    accs_dists.append(accuracy_dists)
```

```python
plt.figure(figsize=(8,6))
plt.plot(emes, delays)
plt.scatter(emes, delays)
plt.grid()
plt.xlabel('Tamao de ventana (m) [muestras]', fontsize=16)
plt.xticks(emes)
plt.ylabel('Retardo de clasificacin [s]', fontsize=16)
plt.title('Retardo de clasificacin vs. Tamao de ventana', fontsize=16)
plt.savefig('grafs/delay_vs_m.eps', format='eps')

plt.figure(figsize=(8,6))
plt.plot(emes, accs_faults)
plt.scatter(emes, accs_faults)
plt.grid()
plt.xlabel('Tamao de ventana (m) [muestras]', fontsize=16)
plt.xticks(emes)
plt.ylabel('Prec. de clasif. de tipo de falla', fontsize=16)
plt.title('Prec. de clasif. de tipo de falla vs. Tamao de ventana', fontsize=16)
plt.savefig('grafs/delay_facc_vs_m.eps', format='eps')

plt.figure(figsize=(8,6))
plt.plot(emes, accs_dists)
plt.scatter(emes, accs_dists)
plt.grid()
plt.xlabel('Tamao de ventana (m) [muestras]', fontsize=16)
plt.xticks(emes)
plt.ylabel('Prec. de clasif. (dist. de insercin)', fontsize=16)
plt.title('Prec. de clasif. (dist. de insercin) vs. Tamao de ventana', fontsize=16)
plt.savefig('grafs/delay_dacc_vs_m.eps', format='eps')
```
